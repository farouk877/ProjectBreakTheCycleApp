```python
items = {}

for i in data['Lineitem name']:
    if i in items.keys():
        items[i] += 1
    else:
        items[i] = 1
items
```

```python
items = data['Lineitem name']
tip = 0
blankOrders = {}
for i in range(len(items)):
    if(data['Lineitem name'][i] == 'Tip'):
        tip += data['Lineitem price'][i]
    else:
        blankRaw = data['Lineitem name'][i].split(' ')
        
        if blankRaw[2] == 'T-Shirt':
            if blankRaw[4] == 'Light':
                shirt = blankRaw[2]
                color = blankRaw[4] + ' ' + blankRaw[5]
                size = blankRaw[7]
            else:
                shirt = blankRaw[2]
                color = blankRaw[4]
                size = blankRaw[6]
        elif blankRaw[2] == 'Hoodie':
            if blankRaw[4] == 'Deep':
                shirt = blankRaw[2]
                color = blankRaw[4] + ' ' + blankRaw[5]
                size = blankRaw[7]
            else:
                shirt = blankRaw[2]
                color = blankRaw[4]
                size = blankRaw[6]
                
        else:
            shirt = blankRaw[2] + " " + blankRaw[3]
            color = blankRaw[5]
            size = blankRaw[7]
            
        key = shirt + ' '+ color + ' '+ size
        
        if key not in blankOrders.keys():
            blankOrders[key]  = 1
        else:
            blankOrders[key] += 1

        
      
        
```
